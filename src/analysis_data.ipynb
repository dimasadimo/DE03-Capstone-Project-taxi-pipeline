```python
import pandas as pd
```

```python

df = pd.read_csv('../staging/staging_data.csv')
df.head(25)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
payment_master = pd.read_csv('../data/payment_type.csv')
df_new = df.merge(payment_master, on="payment_type", how="left", validate="many_to_one")
df_new.rename(columns={"description": "payment_description"}, inplace=True)  # Rename new column
cols = df_new.columns.tolist()
payment_type_index = cols.index('payment_type')
payment_description_index = cols.index('payment_description')
payment_description_col = cols.pop(payment_description_index)
cols.insert(payment_type_index, payment_description_col)
df_new = df_new[cols]
df_new.drop(columns=['payment_type'], inplace=True)
df_new.rename(columns={"payment_description": 'payment_type'}, inplace=True)
df_new.head()
```

```python
df_new["lpep_pickup_datetime"] = pd.to_datetime(df_new["lpep_pickup_datetime"])
df_new["lpep_dropoff_datetime"] = pd.to_datetime(df_new["lpep_dropoff_datetime"])
df_new["trip_duration"] = (df_new["lpep_dropoff_datetime"] - df_new["lpep_pickup_datetime"])

df_new.info()

df_new.head(20)
```

```python
df_new.columns = df_new.columns.str.replace(r'(?<=[a-z0-9])([A-Z])', r'_\1', regex=True).str.replace(r'(?<!^)(?=[A-Z][a-z])', '_', regex=True).str.lower()
df_new.columns
df_new.head(5)

```

```python
df_new['trip_distance'] = (df_new['trip_distance'] * 1.60934).round(2)
df_new.head(5)
```
